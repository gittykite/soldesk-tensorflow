```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import copy

%matplotlib inline  
```

```python
data = [[0, 79],
        [2, 81],
        [4, 93],
        [6, 91],
        [8, 97]] 

studyHr = [i[0] for i in data] 
score = [i[1] for i in data] 

print(type(studyHr))
print('study Hour = ' + str(studyHr))
print('score = ' + str(score))

plt.plot(studyHr, score)
plt.grid(True)
plt.show()
```

```python
np.random.seed(123)

# score = (studyHr * 2) + 50
a = 2
b = 50

studyHr = [i for i in range(6)] 
scorePredict = [(a * hr) + b for hr in studyHr] 
score = [(a * hr) + b + np.random.randint(10) for hr in studyHr] 

print('studyHr = ' + str(studyHr))
print('scorePredict = ' + str(scorePredict))
print('score = ' + str(score))

plt.plot(studyHr, scorePredict, color='black', linestyle='solid', marker = '^')
plt.plot(studyHr, score, color='red', linestyle='solid', marker = 'o')
plt.grid(True)
plt.show()

for i in range(6):
    fmt = 'Study Hour={0}, Predicted Score={1}, Actual Score={2}, Difference={3}'
    print(fmt.format(studyHr[i], scorePredict[i], score[i], (scorePredict[i] - score[i])))

diff = np.array(scorePredict) - np.array(score)
print('diff = ' + str(diff))
```

```python
np.random.seed(123)

# score = (studyHr * 3.3) + 80
a = 3.3
b = 80

studyHr = [i for i in range(6)] 
scorePredict = [(a * hr) + b for hr in studyHr] 

# deepcopy(): new memory assign 
score = copy.deepcopy(scorePredict)
for i in range(len(studyHr)):
    score[i] += np.random.randint(10) - 5

print('studyHr = ' + str(studyHr))
print('scorePredict = ' + str(scorePredict))
print('score = ' + str(score)) 
    
diff = np.array(scorePredict) - np.array(score)
diffSum = np.sum(diff)
diffAvg = round(diffSum / len(diff))

print('diff = ' + str(diff)) 
print('diffSum = ' + str(diffSum)) 
print('diffAvg = ' + str(diffAvg)) 

plt.plot(studyHr, scorePredict, color='black', linestyle='solid', marker = '^')
plt.plot(studyHr, score, color='red', linestyle='solid', marker = 'o')
plt.grid(True)
plt.show()
```

```python
# RMSE: root mean ssquare error
def rmse(p, r):
    diffSq = (p - r) ** 2
    avgDiffSq = diffSq.mean()
    return np.sqrt(avgDiffSq)

print('rmse = %d' % rmse(np.array(scorePredict), np.array(score)))

np.amin(score)
np.amax(score)
np.quantile(score, q=0.5)
np.median(score)
np.average(score)
np.mean(score)
np.std(score)
np.var(score)
np.corrcoef(score, studyHr)
np.correlate(score, studyHr)
np.cov(score, studyHr)

print('amin = %d' % np.amin(score))
print('amax = %d' % np.amax(score))
print('quantile = %d' % np.quantile(score, q=0.5))
print('median = %d' % np.median(score))
print('average = %d' % np.average(score))
print('mean = %d' % np.mean(score))
print('std = %d' % np.std(score))
print('var = %d' % np.var(score))
print('corrcoef = %s' % str(np.corrcoef(score, studyHr)))
print('correlate = %s' % np.array2string(np.correlate(studyHr, score)))
print('cov = %s' % np.array2string(np.cov(score, studyHr)))
```
