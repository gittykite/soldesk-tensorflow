```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

%matplotlib inline  
```

```python
# Step function
def step_function(x):
    return np.where(x < 0, 0, 1)
    # return np.where(x < 0, -1, 1)

# devide -5~5 into 50 points     
x = np.linspace(-5, 5, 50)
print(x)
y = step_function(x)

plt.plot(x, y)
plt.show()
```

```python
# Sigmoid function
#: may cause vanishing gradient problem => use relu instead
def sigmoid_function(x):
    return 1 / (1 + np.exp(-x))

x = np.linspace(-5, 5, 50)
print(x)
y = sigmoid_function(x)

plt.plot(x, y)
plt.show()
```

```python
# Relu Function
def relu_function(x):
    return np.where(x <= 0, 0, x)

x = np.linspace(-5, 5, 50)
print(x)
y = relu_function(x)

plt.plot(x, y)
plt.show()
```

```python
x = np.linspace(-5, 5, 50)
print(x)
y = x

plt.plot(x, y)
plt.show()
```

```python
def softmax_function(x):
    return np.exp(x) / np.sum(np.exp(x))

x = np.array([1, 2, 3])
print(x)
y = softmax_function(x)

print(y)
print('Total:', np.sum(y))
plt.plot(x, y)
plt.show()
```

```python
#  Tanh (hyperbolic tangent) Function
# ex) word combination, time-serise data, RNN
def tanh_function(x):
    return np.tanh(x)

x = np.linspace(-5, 5, 50)
y = tanh_function(x)

plt.plot(x, y)
plt.show()
```
