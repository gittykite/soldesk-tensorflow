### Get image path list

```python
import os

names = ['Choo Mi-ae', 'Kang Kyung-hwa', 'Kersti Kaljulaid', 'Lee Jeongmi', 'Sim Sang-jung', 'Theresa May', 'Tsai Ying-wen']
folder = './src/'
img_path_list = []

for name in names:
    path = folder + name
    file_list = os.listdir(path)
    file_list_py = [file for file in file_list if file.endswith(".jpg")]
    
    for file in file_list_py:
        img_path_list.append(path + '/' + format(file))
        
print('img_path_list=', img_path_list)
```
