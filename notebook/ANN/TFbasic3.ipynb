```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt

# tensorflow 2.0
from tensorflow.keras.models import Sequential  # class
from tensorflow.keras.models import load_model  # model 사용 함수
from tensorflow.keras.layers import Dense       # class

# tensorflow 1.x
# from keras.models import Sequential  # class
# from keras.models import load_model  # model 사용 함수
# from keras.layers import Dense       # class

print(tf.__version__)
```

```python
# Prepare Data
x_train = np.array([[1, 5, 10], [2, 5, 10], [3, 5, 10], [4, 5, 10], [5, 5, 10],
                    [6, 5, 10], [7, 5, 10], [8, 5, 10], [9, 5, 10], [10, 5, 10]])

print(x_train.shape)
print(x_train[0:5])

y_train = np.array([[50], [100], [150], [200], [250], [300], [350], [400], [450], [500]])
print(y_train.shape)

print(y_train.shape)
print(y_train[0:5])
```

```python
# Create Model
model = Sequential()
model2 = Sequential()
model3 = Sequential()

# Single-Layer Perceptron
model.add(Dense(1, input_dim=3, activation='linear'))

# Multi-Layer Perceptron
model2.add(Dense(10, input_dim=3, activation='linear'))
model2.add(Dense(1, activation='linear'))

# Multi-Layer/network Perceptron
model3.add(Dense(30, input_dim=3, activation='linear'))
model3.add(Dense(15, activation='linear'))
model3.add(Dense(6, activation='linear'))
model3.add(Dense(1, activation='linear'))


# Setting Model
# back propagation: adam
# loss algorithm: mse
model.compile(optimizer='adam', loss='mse')
model2.compile(optimizer='adam', loss='mse')
model3.compile(optimizer='adam', loss='mse')
```

```python
model.summary()

# Start Learning (single-layer)
hist = model.fit(x_train, y_train, validation_split=0.2, shuffle=True, epochs=200, batch_size=1)
```

```python
model2.summary()

# Start Learning (multi-layer)
hist2 = model2.fit(x_train, y_train, validation_split=0.2, shuffle=True, epochs=200, batch_size=1)
```

```python
model3.summary()
hist3 = model3.fit(x_train, y_train, validation_split=0.2, shuffle=True, epochs=200, batch_size=1)
```

```python
# Chart Model Result
fig, loss_ax = plt.subplots()

# set y-axis
loss_ax.plot(hist.history['loss'], 'y', label='train loss')
loss_ax.plot(hist.history['val_loss'], 'r', label='val loss')
loss_ax.set_ylim([0.0, 250000.0]) 

# set labels
loss_ax.set_xlabel('epoch')  
loss_ax.set_ylabel('loss')   
loss_ax.legend(loc='upper left')

plt.show()
```
