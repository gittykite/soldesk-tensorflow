```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt

from tensorflow.keras.models import Sequential  
from tensorflow.keras.models import load_model 
from tensorflow.keras.layers import Dense  

%matplotlib inline  
print(tf.__version__)
```

```python
x_train = np.array([[1, 5], [2, 6], [3, 7], [4, 8], [5, 9], [6, 10], [7, 11], [8, 12], [9, 13], [10, 14]])
y_train = np.array([[2, 10], [4, 12], [6, 14], [8, 16], [10, 18], [12, 20], [14, 22], [16, 24], [18, 26], [20, 28]])

print(x_train.shape)
print(y_train.shape)
```

```python
# Initialize weight/bias manually
# tf.keras.initializers.RandomNormal(mean=0.0, stddev=1, seed=None)
# tf.keras.initializers.RandomUniform(minval=-0.05, maxval=0.05, seed=0)
#tf.keras.initializers.RandomUniform(minval=-1.0, maxval=1.0, seed=None)
tf.keras.initializers.RandomNormal(mean=0.0, stddev=0.05, seed=0)

# Create Model
model = Sequential()

# Multi-Layer/network Perceptron
model.add(Dense(60, input_dim=2, activation='linear'))
model.add(Dense(80, activation='linear')) 
model.add(Dense(40, activation='linear')) 
model.add(Dense(15, activation='linear')) 
model.add(Dense(2, activation='linear')) 

# Setting Model
model.compile(optimizer='adam', loss='mse')
```

```python
model.summary()
hist = model.fit(x_train, y_train, validation_split=0.2, shuffle=True, epochs=200, batch_size=1)
```

```python
# Chart Model Result
fig, loss_ax = plt.subplots()

# set y-axis
loss_ax.plot(hist.history['loss'], 'y', label='train loss')
loss_ax.plot(hist.history['val_loss'], 'r', label='val loss')

# set labels
loss_ax.set_xlabel('epoch')  
loss_ax.set_ylabel('loss')   
loss_ax.legend(loc='upper left')

plt.show()
```

```python
x = np.array([[11, 19], [12, 20], [13, 21], [14, 22], [15, 23]])
y = np.array([[22, 38], [24, 40], [26, 42], [28, 44], [30, 46]])

pd = model.predict(x)
print(pd.shape)

for i in range(len(x)):
    fmt = 'x: {0}, actual: {1}, predicted: {2:.5f} {3:.5f}, modified: {4:.0f} {5:.0f}'
    print(fmt.format(x[i], y[i], pd[i][0], pd[i][1], pd[i][0], pd[i][1]))    
    
plt.scatter(x, y, color='g')
plt.plot(x, y, color='g')
plt.scatter(x, pd, color='r')
plt.plot(x, pd, color='r')
plt.grid(True)
plt.show()    
```
