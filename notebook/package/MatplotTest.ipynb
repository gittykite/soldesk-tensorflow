```python
x = [10, 20, 30, 40, 50]
y = [2, 6, 8, 12, 20]

plt.figure(figsize=(8, 3))  
plt.plot(x, y, label='Price')
# x_start, y_start_, x_end, y_end
plt.axis([0, 6, 0, 22000])
plt.title('sales')
plt.xlabel('x product')
plt.ylabel('y amount')
plt.legend()
plt.show()
```

```python
# Random Data
# normalized graph: avg, sigma, num of data
xdata = np.random.normal(5, 3, 100) 
ydata = np.random.normal(3, 5, 100)

print(xdata)
print(type(xdata))

plt.scatter(xdata, ydata, marker='o', s=100)
plt.grid(True)
plt.show()
```

```python
### Linear Regression graph (y= ax + b)

x = [i for i in range(8)]
a = 0
y = [a * i for i in x]

print(x)
print(y)

plt.figure(figsize=(8,3))
plt.plot(x, y, color='blue', linestyle='solid', marker='>')
plt.show()
```

```python
x = [i for i in range(8)] 

a = 1
y = [(a * i) + 1 for i in x]
plt.plot(x, y, color='red', linestyle='solid', marker='+')

a = 2
y = [a * i for i in x]
plt.plot(x, y, color='gold', linestyle='solid', marker='^')

a = 3 
y = [(a * i) + 4 for i in x]
plt.plot(x, y, color='blue', linestyle='solid', marker='h')

plt.grid(True)
plt.show()
```

```python
x = [i for i in range(0, 100, 10)]
print(x)

a = 3
b = 3
c = 10
y = [(a * (i**2)) + (b * i) + c for i in x]
plt.plot(x, y, color='red', linestyle='solid', marker='o')

a = 6
b = 6
c = 20
y = [(a * (i**2)) + (b * i) + c for i in x]
plt.plot(x, y, color='green', linestyle='solid', marker='o')

a = 9
b = 9
c = 30
y = [(a * (i**2)) + (b * i) + c for i in x]
plt.plot(x, y, color='blue', linestyle='solid', marker='o')

plt.grid(True)
plt.show()
```
